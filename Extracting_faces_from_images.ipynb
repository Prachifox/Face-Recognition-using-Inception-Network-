```python
import cv2
import dlib
import matplotlib.pyplot as plt
import matplotlib.patches as patches

from align import AlignDlib

%matplotlib inline


def load_image(path):
    img = cv2.imread(path, 1)
    # OpenCV loads images with color channels
    # in BGR order. So we need to reverse them
    return img[:,:,[2,1,0]]

# Initialize the OpenFace face alignment utility
alignment = AlignDlib('models/landmarks.dat') #detects face in image and also shape of face

# Load an image of Jacques Chirac
img=cv2.imread('IMG-20160821-WA0016.jpg')
plt.subplot(141)
plt.imshow(img)
# Detect face and return bounding box
bbs = alignment.getAllFaceBoundingBoxes(img)
count=0
k=142
for bb in bbs:
    count=count+1
    alignedFace=alignment.align(96,img,bb,landmarkIndices=AlignDlib.OUTER_EYES_AND_NOSE)
    plt.subplot(k)
    plt.imshow(alignedFace)
    k=k+1
print(count)    
    
```
